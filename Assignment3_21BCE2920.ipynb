# **Handling Null Values**
As, we can see there are null values present in the dataset in the Age, Cabin and Embarked

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

We will therefore use mean and mode to replace the null values.
We shall use the following codes because age is a numerical column and cabin and embarked are categorical columns. Now we can conduct another check to eliminate any null values.

**IMPUTATION WITH MEAN FOR AGE COLUMN**

```python
df["Age"].fillna(df["Age"].mean(),inplace=True)
```

```python
df["Age"].isnull().sum()
```

```python
df["Cabin"].fillna(df["Cabin"].mode()[0],inplace=True)
```

```python
df["Embarked"].fillna(df["Embarked"].mode()[0],inplace=True)
```

# **Outlier Detection**

Here,  for passenger Id there are no outliers, but for the Age, we have the outliers, so in order to remove the outliers, I have used the IQR method

```python
sns.boxplot(df["PassengerId"])
```

```python
sns.boxplot(df.Age)
```

```python
q1 = df.Age.quantile(0.25) #qi mean 25 percentage of data
q3 = df.Age.quantile(0.75)
```

```python
IQR = q3-q1
IQR
```

```python
upper_limit = q3+1.5*IQR
```

```python
upper_limit
```

```python
df.median() #50% quantile
```

```python
df = df[df.Age<upper_limit]
```

```python
sns.boxplot(df.Age)
```

```python
sns.boxplot(df.Pclass)
```

```python
sns.boxplot(df.Parch)
```

```python
lower_percentile = 5
upper_percentile = 85

# Calculate the lower and upper bounds based on percentiles
lower_bound = df['Age'].quantile(lower_percentile / 100)
upper_bound = df['Age'].quantile(upper_percentile / 100)
```

```python
df_no_outliers = df[(df['Age'] >= lower_bound) & (df['Age'] <= upper_bound)]

# Create a boxplot to visualize the 'Age' column without outliers
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_no_outliers, y='Age')
plt.title('Boxplot of Age without Outliers')
plt.show()
```
